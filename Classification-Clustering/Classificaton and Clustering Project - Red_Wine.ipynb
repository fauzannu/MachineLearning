## **2) Regression** : Prediksi berapa kualitas red wine berdasarkan komposisinya?

### **Multiple Linear Regression**

```python
X = red_wine[['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar', 'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density', 'pH', 'sulphates', 'alcohol']]
y = red_wine['quality']
X.head()
```

#### **Split the dataset**

```python
#Create train & test dataset
from sklearn.model_selection import train_test_split

x_train, x_test,y_train,y_test = train_test_split(X,y,test_size =0.2, random_state = 10) #test-size 0,2 = test data = 20%, train data = 80%, random_state supaya randomnya gaberubah2
print("X_train shape: ",x_train.shape)
print("X_test shape: ",x_test.shape)
print("y_train shape: ",y_train.shape)
print("y_test shape: ",y_test.shape)
```

#### **Fit the model**

```python
from sklearn import linear_model

regr = linear_model.LinearRegression()
regr.fit(x_train, y_train)
```

#### **Predict the output**

```python
predict_quality = regr.predict(x_test)
print(predict_quality)
```

#### **Evaluation**

```python
print('R-square: ', regr.score(X, y))
print ('Intercept: ',regr.intercept_)
print ('Coefficients: ', regr.coef_)
```

```python
from sklearn.metrics import mean_squared_error
from math import sqrt

Y_predict_multifit = regr.predict(X)

print('RMSE:', sqrt(mean_squared_error(red_wine['quality'], Y_predict_multifit)))
print('MSE:', mean_squared_error(red_wine['quality'], Y_predict_multifit))
```

### **Regression**

```python
## LINEAR REGRESSION

from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn.metrics import mean_squared_error, mean_absolute_error
from math import sqrt

def reg_evaluation(y_true, y_pred):
  print('MAE :', mean_absolute_error(y_true, y_pred))
  print('MSE :', mean_squared_error(y_true, y_pred))
  print('RMSE:', sqrt(mean_squared_error(y_true, y_pred)))

X_sel = red_wine[['fixed acidity', 'volatile acidity', 'citric acid', 'sulphates', 'alcohol']]

X_train, X_test, Y_train, Y_test = train_test_split(X_sel, y, test_size =0.2, random_state = 26)



sc = StandardScaler().fit(X_train)

X_train_std = sc.transform(X_train)
X_test_std = sc.transform(X_test)

regr = linear_model.LinearRegression()
regr.fit(X_train_std, Y_train)

Y_pred = regr.predict(X_test_std)
Y_train_pred = regr.predict(X_train_std)

print('========== Test dataset score ==========')
print('R-square: ', regr.score(X_test_std, Y_test))
reg_evaluation(Y_test, Y_pred)

print('\n========== Train dataset score ==========')
print('R-square: ', regr.score(X_train_std, Y_train))
reg_evaluation(Y_train, Y_train_pred)
```

### **Polynomial Features**

```python
## POLYNOMIAL FEATURES
print('========== Polynomial Features ==========')

poly_features = PolynomialFeatures(degree=3).fit(X_train)
X_train_poly = poly_features.transform(X_train)
X_test_poly = poly_features.transform(X_test)

sc2 = StandardScaler().fit(X_train_poly)
X_train_poly = sc2.transform(X_train_poly)
X_test_poly = sc2.transform(X_test_poly)

regr_poly = linear_model.LinearRegression()
regr_poly.fit(X_train_poly, Y_train)

Y_poly_pred = regr_poly.predict(X_test_poly)
Y_train_poly_pred = regr_poly.predict(X_train_poly)

print('========== Test dataset score ==========')
print('R-square: ', regr_poly.score(X_test_poly, Y_test))
reg_evaluation(Y_test, Y_poly_pred)

print('\n========== Train dataset score ==========')
print('R-square: ', regr_poly.score(X_train_poly, Y_train))
reg_evaluation(Y_train, Y_train_poly_pred)
```

**Catatan**: Dari percobaan penggunaan model Multiple Linear Regression (MLR) dan Polynomial Regression, R-square score pada Polynomial Regression lebih besar daripada score hasil percobaan pada MLR.

Nilai RMSE & MSE pada Polynomial Regression juga lebih kecil dibandingkan dengan nilai RMSE & MSE pada MLR. Sehingga model Polynomial Regression lebih cocok dalam memprediksi kualitas red wine dengan teknik regresi.

For loop model Clustering

```python
from sklearn.metrics import silhouette_score

def train_cluster(model, X, name):

    silhouette_scores = []
    list_k = list(range(2, 15))

    for k in list_k:
        clust = model(k)
        clusters = clust.fit_predict(X)
        silhouette_scores.append(silhouette_score(X, clusters))

    clust = model(np.argmax(silhouette_scores)+2)
    clusters = clust.fit_predict(X)
    
    X_clu = X.copy()
    X_clu['cluster'] = clusters

    print('Silhouette Scores ({}) :'.format(name), max(silhouette_scores))

    # Plot sse against k
    plt.figure(figsize=(15, 6))
    plt.suptitle(name)

    plt.subplot(1, 2, 1)
    plt.plot(list_k, silhouette_scores, '-o')
    plt.xlabel(r'Number of clusters *k*')
    plt.ylabel('Silhouette score')

    plt.subplot(1, 2, 2)
    sns.scatterplot(data=X_clu, x="PC1", y="PC2", hue="cluster")
    # sns.scatterplot(data=pd.DataFrame(clust.cluster_centers_, columns=['C1', 'C2']), x='C1', y='C2', s=300, color='red')

    plt.show()
    
    return clusters
```

```python
from sklearn.decomposition import PCA
from sklearn.mixture import GaussianMixture
from sklearn.cluster import KMeans, SpectralClustering, MiniBatchKMeans, AgglomerativeClustering

X_cl = red_wine.drop(['quality'], axis=1)
reduced_X = PCA(n_components=2).fit_transform(X_cl)
reduced_X = pd.DataFrame(StandardScaler().fit_transform(reduced_X), columns=["PC1", "PC2"])

kmeans = train_cluster(KMeans, reduced_X, 'K-Means Clustering')
gaussian = train_cluster(GaussianMixture, reduced_X, 'GaussianMixture Clustering')
spectral = train_cluster(SpectralClustering, reduced_X, 'SpectralClustering Clustering')
meanshift = train_cluster(MiniBatchKMeans, reduced_X, 'MiniBatchKMeans Clustering')
algome = train_cluster(AgglomerativeClustering, reduced_X, 'AgglomerativeClustering Clustering')
```
